#### 2-1. PreProcessing: Popularity classifier

Before running the logistic regression, we turned actual popularity score to a popular level. In order to do classification, we converted `popularity` to `popularity_type`.

According to the general standards by Spotify, popularity could be categorized as:

> **Very Popular** : score higher than 70.   
> **Somewhat Popular** : score between 35 and 70.   
> **Unpopular** :score below 35. (50% observations of our dataset)

```python
# create new column
df_clean['popularity_type']=" "

def function(a):
  if a > 70: return 3
  elif a > 35:return 2
  else:return 1

# classify the popularity score into pre-defined types
df_clean['popularity_type'] = df_clean.apply(lambda x:function(x['popularity']),axis=1)
df_clean['popularity_type'] = df_clean['popularity_type'].replace(3, "Very Popular").replace(2, "Somewhat Popular").replace(1, "Unpopular")

# change column type into category
df_clean['popularity_type'] = df_clean['popularity_type'].astype('category')
df_clean.head(3)
```

###### 2-1-1. PreProcessing: Construct Pipeline

Though there is only one row of missing value and was deleted in the data cleaning phase, we still keep the imputer in our pipeline. Then, we choose "StandardScaler" for numerical variable and "OneHotEncode" for categorical varibles. We did consider using other scale encodings, but there's no sequence realtionship in our categorical variables. Hence, we stil used One-Hot to  assign new binary feature for each category feature.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn import set_config

set_config(display='diagram') # shows the pipeline structure graphically

num_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ])
from sklearn.preprocessing import OneHotEncoder
cat_pipeline = Pipeline([
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("cat_encoder", OneHotEncoder(sparse=False, drop="first"))
    ])

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

num_attribs = ['duration_min','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']
cat_attribs = ["track_genre", "explicit",'key','mode','time_signature']

# The following step patches SimpleImputer to produce feature names
SimpleImputer.get_feature_names_out = StandardScaler.get_feature_names_out

preprocess_pipeline = ColumnTransformer([ # handle each type of column with appropriate pipeline
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline, cat_attribs),
    ])

preprocess_pipeline
```

##### 2.9.1 Over Sampling (Accuracy:0.71)

Note: if average='micro', then precision/ recall/ f1 scores can be equal. Its given in the user guide:

Note that for “micro”-averaging in a multiclass setting with all labels included will produce equal precision, recall and F, while “weighted” averaging may produce an F-score that is not between precision and recall.

We used 'weighted' at this point.

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.pipeline import Pipeline, make_pipeline
from sklearn.linear_model import LogisticRegression

lr_under_pipe = make_pipeline(preprocess_pipeline, RandomOverSampler(random_state=42), LogisticRegression())
lr_under_pipe.fit(df_X, y_train)
y_pred = lr_under_pipe.predict(df_tX)
print_scores(y_test, lr_under_pipe.predict(df_tX));
print("Accuracy:",accuracy_score(y_test, y_pred))
print("Balanced Accuracy: ",balanced_accuracy_score(y_test, y_pred))
print("Precision Score: ",precision_score(y_test, y_pred,average='weighted'));
print("Recall Score: ",recall_score(y_test, y_pred, average='weighted'));
print("F1 Score: ", f1_score(y_test, y_pred,average='weighted'))  
```

## Conclusion

Our team aims to evaluate machine learning models to predict Spotify song popularity and identify what makes a song popular. We performed data cleaning, exploration of the dataset, and build ML models in Python.

In phase one, we implemented exploratory data analysis. For better data usage, we changed the data types, deleted one missing value, and dealt with outliers in the "tempo" column.
